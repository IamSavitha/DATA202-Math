```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as interpolate
from scipy.interpolate import Rbf
```

```python
np.random.seed(6)
kinds = ('nearest', 'zero', 'linear', 'slinear', 'quadratic', 'cubic')
```

```python
N = 10
x = np.linspace(0, 1, N)
y = np.random.randint(10, size=(N,))
```

```python
new_x = np.linspace(0, 1, 28)
fig, axs = plt.subplots(nrows=len(kinds)+3, sharex=True, figsize=[10,20])
axs[0].plot(x, y, 'bo-')
axs[0].set_title('raw')

for ax, kind in zip(axs[1:], kinds):
    new_y = interpolate.interp1d(x, y, kind=kind)(new_x)
    ax.plot(new_x, new_y, 'ro-')
    ax.set_title(kind)
    
new_y = interpolate.lagrange(x,y)(new_x)
axs[7].plot(new_x, new_y, 'ro-')
axs[7].set_title('Lagrange')

new_y = interpolate.Rbf(x, y)(new_x)
axs[8].plot(new_x, new_y, 'ro-')
axs[8].set_title('Rbf')

plt.show()
```
