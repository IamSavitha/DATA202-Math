```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from scipy import interpolate

x = np.arange(-2.01, 2.01, 0.25)
y = np.arange(-2.01, 2.01, 0.25)
xx, yy = np.meshgrid(x, y)

z = np.sin(xx)**2 + np.cos(yy)**2
f = interpolate.interp2d(x, y, z, kind='linear')

xnew = np.arange(-2.01, 2.01, 0.1)
ynew = np.arange(-2.01, 2.01, 0.1)
xxnew, yynew = np.meshgrid(xnew, ynew)

kinds = ('linear', 'cubic', 'quintic')
f = plt.figure(figsize=(10,10))

axs = f.add_subplot(2,2,1, projection='3d')
axs.scatter(xx, yy, z)
axs.set_title('raw')

for kind, ind in zip(kinds, range(3)):
    znew = interpolate.interp2d(x, y, z, kind=kind)(xnew, ynew)
    ax = f.add_subplot(2,2, ind+2, projection='3d')
    ax.plot_surface(xxnew, yynew, znew, cmap=cm.coolwarm, linewidth=0, antialiased=False)
    ax.set_title(kind)
    
plt.show()
```
