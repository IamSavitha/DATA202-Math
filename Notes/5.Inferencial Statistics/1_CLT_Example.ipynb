
# Central Limit Theorem (CLT) Example

This notebook provides an example to demonstrate the Central Limit Theorem (CLT). 
The CLT states that the sampling distribution of the sample means will approximate a normal distribution as the sample size becomes larger, regardless of the population's distribution shape.

## Example Problem:
Suppose we have a population with a non-normal distribution (e.g., a uniform distribution). We will draw multiple samples from this population and compute their means. According to the CLT, as the sample size increases, the distribution of these sample means should approximate a normal distribution.

Let's explore this with a practical example:


```python

import numpy as np
import matplotlib.pyplot as plt

# Create a population with a uniform distribution
population = np.random.uniform(low=0, high=20, size=10000)
plt.hist(population, bins=30, edgecolor='black', color='yellow')
plt.title(f"population Distribution ")
plt.xlabel("age")
plt.ylabel("count")
plt.show()
    
# Function to plot histograms
def plot_sample_means(sample_size, num_samples=1000):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))
    
    plt.hist(sample_means, bins=30, edgecolor='black')
    plt.title(f"Sample Means Distribution (Sample Size = {sample_size})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")
    plt.show()

# Plot for different sample sizes
plot_sample_means(sample_size=5)
plot_sample_means(sample_size=30)
plot_sample_means(sample_size=100)
plot_sample_means(sample_size=500)
plot_sample_means(sample_size=1000)
```
